# Actividad 1: Eliminación de Duplicados y Tratamiento de Datos

```python
import pandas as pd
import seaborn as sns
# Cargar el dataset de Titanic
df = sns.load_dataset('titanic')
print("Primeros registros")
print(df.head())

# 1. Filtrar pasajeros sobrevivientes
# Usamos la columna 'alive' para obtener solo los registros que sean 'yes'
sobrevivientes = df[df['alive'] == 'yes'].copy()

# 2. Seleccionar columnas relevantes
# Filtramos el DataFrame de sobrevivientes para quedarnos solo con sex, age y fare
analisis_sobrevivientes = sobrevivientes[['sex', 'age', 'fare']]

# 3. Crear una nueva columna de categoría
# Para este paso, aprovechamos la columna 'class' (o 'pclass') que suele venir en este dataset.
# La convertimos explícitamente al tipo 'category' para optimizar el análisis.
analisis_sobrevivientes['categoria'] = df['class'].astype('category')

# Visualizar el resultado
print("Resultados")
print(analisis_sobrevivientes.head())
```

# Actividad 2: Manipulación de Datos

```python
import pandas as pd
import seaborn as sns
# Cargar el dataset de Titanic
df = sns.load_dataset('titanic')
print("Primeros registros")
print(df.head())

# 1. Obtener los nombres de las columnas
nombres_columnas = df.columns
print("Columnas actuales:", nombres_columnas)

# 2. Eliminar la columna 'deck'
# Usamos errors='ignore' para evitar errores si la columna ya no existe
df_limpio = df.drop(columns=['deck'], errors='ignore')

# 3. Reindexar el DataFrame
# Al eliminar una columna no se desordenan las filas, pero si realizaste
# filtrados previos, reset_index asegura que el conteo sea continuo.
df_reindexado = df_limpio.reset_index(drop=True)

# Verificación
print(df_reindexado.head())

```

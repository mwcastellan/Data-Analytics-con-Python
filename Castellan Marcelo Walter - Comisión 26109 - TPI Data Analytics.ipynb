## 1- Cargar Datasets

```python
def cargar_dataset(link):
  # Usamos regex para extraer el file_id del link
  ahora = datetime.now()
  print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
  file_id = link.split('/d/')[1].split('/')[0]

  # Con ese file_id generamos la url
  url = f"https://drive.google.com/uc?id={file_id}"

  # Finalmente lo importamos usando el método read_csv() con la url generada
  df = pd.read_csv(url)
  ahora = datetime.now()
  print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
  return df
```

## 4- Analizar Datos

```python
def analizar_datos(df):
  ahora = datetime.now()
  print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

  print("📊 Dimensiones del dataset")
  print(f"Filas: {df.shape[0]}")
  print(f"Columnas: {df.shape[1]}")
  print("-" * 40)

  print("\n🔎 Tipos de datos")
  print(df.dtypes)
  print("-" * 40)

  print("\n🔎 Primeros registros")
  print(df.head(2))
  print("-" * 40)

  print("\n📈 Estadísticas descriptivas")
  print(df.describe(include='number'))
  print("-" * 40)

  print("\n📈 Estadísticas descriptivas")
  try:
        print(df.describe(include='object'))
  except:
      print("No hay variables categóricas")
  print("-" * 40)

  print("\n⚠️ Valores nulos por columna")
  nulos = df.isnull().sum()
  if (nulos > 0).any():
        print(nulos[nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos en el DataFrame (0 en todas las columnas)")
  print("-" * 40)

  print("\n📌 Porcentaje de nulos por columna")
  porcentaje_nulos = df.isnull().mean() * 100
  if (porcentaje_nulos > 0).any():
        print(porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos (0% en todas las columnas)")
  print("-" * 40)

  print("\n🔁 Registros duplicados")
  duplicados = df.duplicated().sum()
  print(f"Cantidad de duplicados: {duplicados}")
  print("-" * 40)

  ahora = datetime.now()
  print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

# Etapa 2: Preprocesamiento y Limpieza de Datos (Clases 5 a 8)

## 1- Limpieza de datos

```python
def limpiar_datos(df):
  d=df.copy()
  duplicados = d.duplicated().sum()
  print("\n🔁 Registros duplicados")
  print(f"Cantidad de duplicados inicio: {duplicados}")
  d = d.drop_duplicates()
  duplicados = d.duplicated().sum()
  print(f"Cantidad de duplicados fin: {duplicados}")

  print("\n⚠️ Valores nulos inicio por columna")
  nulos = d.isnull().sum()
  if (nulos > 0).any():
        print(nulos[nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos en el DataFrame (0 en todas las columnas)")
  print("-" * 40)
  print("\n📌 Porcentaje de nulos por columna")
  porcentaje_nulos = d.isnull().mean() * 100
  if (porcentaje_nulos > 0).any():
        print(porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos (0% en todas las columnas)")
  print("-" * 40)

  d = d.dropna()
  print("\n⚠️ Valores nulos fin por columna")
  nulos = d.isnull().sum()
  if (nulos > 0).any():
        print(nulos[nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos en el DataFrame (0 en todas las columnas)")
  print("-" * 40)
  print("\n📌 Porcentaje de nulos por columna")
  porcentaje_nulos = d.isnull().mean() * 100
  if (porcentaje_nulos > 0).any():
        print(porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False))
  else:
        print("No hay valores nulos (0% en todas las columnas)")
  print("-" * 40)

  return d


print("-------------D U P L I C A D O S   Y  N U L O S ---------------------")
print("====================== CLIENTES =====================================")
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
df_clientes_l = limpiar_datos(df_clientes)
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

print("====================== VENTAS ======================================")
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
df_ventas_l = limpiar_datos(df_ventas)
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

print("===================== MARKETING ====================================")
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
df_marketing_l = limpiar_datos(df_marketing)
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

## 2- Transformación de Datos (Ventas)

```python
# hacer una copia
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

df_ventas_p = df_ventas_l.copy()
df_ventas_p.info()
print(df_ventas_p.head(5))
# remover $ del precio
df_ventas_p['precio'] = df_ventas_p['precio'].str.replace('$', '')
# convertir precio de object a float
df_ventas_p['precio'] = pd.to_numeric(df_ventas_p['precio'], errors="coerce")
###df_ventas_p["precio"].isnull().sum()
# si quieren que cantidad quede en int (alternativamente)
df_ventas_p['cantidad'] = pd.to_numeric(df_ventas_p['cantidad'], errors='coerce')
df_ventas_p['cantidad'] = df_ventas_p['cantidad'].fillna(0)
df_ventas_p['cantidad'] = df_ventas_p['cantidad'].astype(int)
df_ventas_p['cantidad'] = pd.to_numeric(df_ventas_p['cantidad'], errors='coerce').fillna(0).astype(int)

# nueva columna de total ventas
df_ventas_p["total_ventas"] = df_ventas_p["precio"] * df_ventas_p["cantidad"]
print(df_ventas_p.head(5))
df_ventas_p.info()

ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

### 2.1 Productos con alto rendimiento

```python
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

df_ventas_r = df_ventas_p.copy()
df_ventas_r.info()
df_ventas_r.sort_values(by='producto', ascending=True).head()

# Hacemos el group por producto y la agregacion por total_ventas (suma)
df_ventas_agg = df_ventas_r.groupby(["producto"], as_index=False)["total_ventas"].sum()
df_ventas_agg.sort_values(by="total_ventas", ascending=True).head()

# Para el alto rendimiento, opcion 1:
# Calculamos el promedio de las sumatorias de ventas de productos
promedio = df_ventas_agg["total_ventas"].mean()
print("Promedio: " + str(promedio))

# Para el alto rendimiento, opcion 2:
# Calculamos percentil del 90% o el que quieran
quantile_90 = df_ventas_agg["total_ventas"].quantile(0.9)
print("Quantile: " + str(quantile_90))

# Alto rendimiento sobre el promedio
print("Alto rendimiento sobre el promedio:")
alto_renimiento_1 = df_ventas_agg[df_ventas_agg["total_ventas"] > promedio]
print(alto_renimiento_1.sort_values(by="total_ventas", ascending=True).head())

# Alto rendimiento sobre el top 10%
print("Alto rendimiento sobre el top 10%:")
alto_renimiento_2 = df_ventas_agg[df_ventas_agg["total_ventas"] > quantile_90]
print(alto_renimiento_2.sort_values(by="total_ventas", ascending=False).head())

# Un gráfico
import matplotlib.pyplot as plt
import seaborn as sns
df_ventas_agg = df_ventas_agg.sort_values(by="total_ventas", ascending=True)
sns.barplot(x="producto", y="total_ventas", data=df_ventas_agg)
plt.xticks(rotation=90)
plt.show()

ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# Obtener el promedio de las ventas promedio por categotia
df_ventas_p.groupby(["categoria"], as_index=False)["total_ventas"].mean()
```

```python
# Agrupando por Category, generar 3 columnas con:
# Suma de ventas como ventas_totales
# Cantidad de ventas como cant_ventas
# Promedio de ventas como avg_ventas
# Validar el promedio
df_ventas_p.groupby(["categoria"], as_index=False).agg(
    ventas_totales=("total_ventas", "sum"),
    cant_ventas=("cantidad", "count"),
    avg_ventas=("total_ventas", "mean")
)

ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
print('* ---***---***---***---***---***---***---***---***---***---***---***---***  *')
print ('C26109 - ENTREGA FINAL PFI')
print('* ------------------------------------------------------------------------- *')
print('ETAPA 3: ANALISIS DE DATOS')
print('* ------------------------------------------------------------------------- *')

print('1.Estadística Descriptiva')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
# Copia de Ventas.
df_ventas = df_ventas_p.copy()
# Corrijo Fecha y agrego Período.
df_ventas['fecha_venta'] = pd.to_datetime(df_ventas['fecha_venta'],dayfirst=True)
df_ventas['año_mes'] = df_ventas['fecha_venta'].dt.to_period('M')
print(df_ventas)
print(df_ventas.dtypes)

# Tendencia central.
wk_ventas = df_ventas['total_ventas']
print('Tendencia central')
print(f"Media: {wk_ventas.mean():.2f}")
print(f"Mediana: {wk_ventas.median():.2f}")
print("Moda:")
print(wk_ventas.mode())

# Dispersión.
print('Dispersión')
print(f"\nDesviación estándar: {wk_ventas.std():.2f}")
print(f"Q1: {wk_ventas.quantile(0.25):.2f}")
print(f"Q3: {wk_ventas.quantile(0.75):.2f}")
print(f"Rango intercuartílico (IQR): {(wk_ventas.quantile(0.75)-wk_ventas.quantile(0.25)):.2f}")
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# ---------------------------------------------------------------------
print('2- EDA (Análisis Exploratorio de Datos)')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
import matplotlib.pyplot as plt
import seaborn as sns
# Graficos por productos y total ventas.
wk_ventas=df_ventas.groupby(["producto"], as_index=False).agg(
    total_ventas=("total_ventas", "sum"),
)
# De barras.
plt.figure(figsize=(14,6))
plt.bar(
    range(len(wk_ventas)),
    wk_ventas['total_ventas'],
    width=0.7,      # barras más angostas.
    color='skyblue'
)
plt.xticks(
    range(len(wk_ventas)),
    wk_ventas['producto'],
    rotation=45,
    ha='right'
)
plt.tight_layout()
plt.title('Ventas totales por producto')
plt.show()

# De distribución de ventas.
sns.histplot(wk_ventas['total_ventas'], bins=4, kde=True)
plt.title('Distribución de ventas')
plt.xlabel('Total_ventas')
plt.ylabel('Frecuencia')
plt.show()

# Graficos periodo y total ventas.
wk_ventas=df_ventas.groupby(df_ventas["año_mes"], as_index=False).agg(
    total_ventas=("total_ventas", "sum"),
)
# De barras.
plt.figure(figsize=(14,6))
plt.bar(
    range(len(wk_ventas)),
    wk_ventas['total_ventas'],
    width=0.7,      # barras más angostas.
    color='skyblue'
)
plt.xticks(
    range(len(wk_ventas)),
    wk_ventas['año_mes'],
    rotation=45,
    ha='right'
)
plt.tight_layout()
plt.title('Ventas totales por periodo')
plt.show()

# Graficos por categoria y total ventas.
wk_ventas=df_ventas.groupby(["categoria"], as_index=False).agg(
    total_ventas=("total_ventas", "sum"),
)
# De barras.
plt.figure(figsize=(14,6))
plt.bar(
    range(len(wk_ventas)),
    wk_ventas['total_ventas'],
    width=0.7,      # barras más angostas.
    color='skyblue'
)
plt.xticks(
    range(len(wk_ventas)),
    wk_ventas['categoria'],
    rotation=45,
    ha='right'
)
plt.tight_layout()
plt.title('Ventas totales por categoria')
plt.show()
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# ---------------------------------------------------------------------
print('3- Correlación')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# Seleccion: Lámpara de mesa, por ser el más vendido.
producto_sel = 'Lámpara de mesa'
print('Producto: ' + producto_sel)
# Agrego fechas como numero para hacer su correlación.
df_ventas['fecha_num'] = df_ventas['fecha_venta'].astype('int64')
df_ventas['periodo_num'] = df_ventas['año_mes'].astype('int64')
# Selección.
wk_ventas = df_ventas[(df_ventas["producto"] == producto_sel)]
print(wk_ventas)
wk_ventas_corr = wk_ventas[
    ['precio','cantidad','total_ventas','fecha_num','periodo_num']
].corr()
print(wk_ventas_corr)

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.heatmap(
    wk_ventas_corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Matriz de Correlación - ' + producto_sel)
plt.show()
sns.scatterplot(x='precio', y='cantidad', data=wk_ventas)
plt.show()

sns.scatterplot(x='precio', y='total_ventas', data=wk_ventas)
plt.show()
sns.scatterplot(x='periodo_num', y='total_ventas', data=wk_ventas)
plt.show()
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# ---------------------------------------------------------------------
print('4.Consolidación de datos')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

# Ventas.
# df_ventas
# Marketing.
df_marketing = df_marketing_l.copy()
# Corrijo Fechas.
df_marketing['fecha_inicio'] = pd.to_datetime(df_marketing['fecha_inicio'],dayfirst=True)
df_marketing['fecha_fin'] = pd.to_datetime(df_marketing['fecha_fin'],dayfirst=True)

# Consolidado.
wk_consolidado = pd.merge(df_ventas, df_marketing, on="producto", how="left")
# Indica si la venta fué realizada en alguna campaña o no.
wk_consolidado["venta_en_campaña"] = (
    (wk_consolidado["fecha_venta"] >= wk_consolidado["fecha_inicio"]) &
    (wk_consolidado["fecha_venta"] <= wk_consolidado["fecha_fin"])
)
print(wk_consolidado.sort_values(by='id_venta', ascending=True))
print(wk_consolidado.dtypes)

df_consolidado = (
    wk_consolidado.sort_values(by='id_venta', ascending=False)
    .groupby("id_venta", as_index=False)
    .agg({
        "producto": "first",
        "precio": "first",
        "cantidad": "first",
        "fecha_venta": "first",
        "año_mes": "first",
        "categoria": "first",
        "total_ventas": "first",
        "venta_en_campaña": "max"
    })
)
print(df_consolidado)
print(df_consolidado.dtypes)
print(
    df_consolidado
    .groupby(['producto', 'venta_en_campaña'])['total_ventas']
    .agg(['count', 'sum', 'mean'])
)
# Aqui se repite id_venta, es para totales por canal marketing
df_consolidado_II = (wk_consolidado.sort_values(by='id_venta', ascending=False))
print(df_consolidado_II.dtypes)
ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
print('* ---***---***---***---***---***---***---***---***---***---***---***---***  *')
print ('C26109 - ENTREGA FINAL PFI')
print('* ------------------------------------------------------------------------- *')
print('ETAPA 4: VISUALIZACION DE DATOS')
print('* ------------------------------------------------------------------------- *')

print('1.Visualización con seaborn: generar un gráfico de histogramas y boxplots de las variables de interés.')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

import seaborn as sns
import matplotlib.pyplot as plt

# Consolidado.
print(df_consolidado.dtypes)

variables_numericas = [
    'precio',
    'cantidad',
    'total_ventas'
]
# Histogramas.
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, var in zip(axes, variables_numericas):
    sns.histplot(
        data=df_consolidado,
        x=var,
        kde=True,
        bins=30,
        ax=ax
    )
    ax.set_title(f'Histograma de {var}')
plt.tight_layout()
plt.show()

# Boxplot.
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, var in zip(axes, variables_numericas):
    sns.boxplot(
        data=df_consolidado,
        x=var,
        ax=ax
    )
    ax.set_title(f'Boxplot de {var}')
plt.tight_layout()
plt.show()

# Boxplot por campañas.
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df_consolidado,
    x='venta_en_campaña',
    y='total_ventas'
)
plt.title('Total de ventas según campaña')
plt.show()

# Histogramas según campaña.
plt.figure(figsize=(8,5))
sns.histplot(
    data=df_consolidado,
    x='total_ventas',
    hue='venta_en_campaña',
    kde=True,
    bins=30,
    alpha=0.5
)
plt.title('Distribución de total_ventas según campaña')
plt.show()

ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

print('2.Visualización con Plotly.')
ahora = datetime.now()
print("Fecha Inicio: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))

import plotly.express as px

# a. Ventas mensuales.
ventas_mes = (
    df_consolidado
    .groupby('año_mes', as_index=False)
    .agg(
        total_ventas=('total_ventas', 'sum'),
        cantidad=('cantidad', 'sum')
    )
)
ventas_mes['año_mes'] = ventas_mes['año_mes'].astype(str)
fig = px.line(
    ventas_mes,
    x='año_mes',
    y='total_ventas',
    markers=True,
    title='Evolución mensual de las ventas'
)
fig.update_layout(
    xaxis_title='Mes',
    yaxis_title='Ventas'
)
fig.show()

# b. Comparar campaña vs no campaña.
ventas_campania = (
    df_consolidado
    .groupby('venta_en_campaña', as_index=False)
    .agg(total_ventas=('total_ventas', 'sum'))
)
fig = px.bar(
    ventas_campania,
    x='venta_en_campaña',
    y='total_ventas',
    color='venta_en_campaña',
    title='Ventas con y sin campaña'
)
fig.show()

# c. Ventas por categoría.
ventas_categoria = (
    df_consolidado
    .groupby('categoria', as_index=False)
    .agg(total_ventas=('total_ventas', 'sum'))
    .sort_values('total_ventas', ascending=False)
)
fig = px.bar(
    ventas_categoria,
    x='categoria',
    y='total_ventas',
    color='total_ventas',
    title='Ventas por categoría'
)
fig.show()

# d. Ventas por producto y campaña.
ventas_producto = (
    df_consolidado
    .groupby(['producto', 'venta_en_campaña'], as_index=False)
    .agg(total_ventas=('total_ventas', 'sum'))
)
fig = px.bar(
    ventas_producto,
    x='producto',
    y='total_ventas',
    color='venta_en_campaña',
    barmode='group',
    title='Ventas por producto según campaña'
)
fig.update_layout(xaxis_tickangle=-45)
fig.show()

# e. Version II
ventas_mes_campania = (
    df_consolidado
    .groupby(['año_mes', 'venta_en_campaña'], as_index=False)
    .agg(total_ventas=('total_ventas', 'sum'))
)
ventas_mes_campania['año_mes'] = ventas_mes_campania['año_mes'].astype(str)
fig = px.line(
    ventas_mes_campania,
    x='año_mes',
    y='total_ventas',
    color='venta_en_campaña',
    markers=True,
    title='Evolución de ventas: campaña vs no campaña'
)
fig.show()

# f. Consolidado II
resumen = (
    df_consolidado_II
    .groupby(['canal','venta_en_campaña'], as_index=False)
    .agg(
        ventas=('total_ventas','sum'),
        transacciones=('id_venta','nunique')
    )
)
fig = px.bar(
    resumen,
    x='canal',
    y='ventas',
    color='venta_en_campaña',
    barmode='group',
    title='Ventas por canal durante campañas'
)
fig.show()

ahora = datetime.now()
print("Fecha Fin: " + ahora.strftime("%Y-%m-%d %H:%M:%S"))
```

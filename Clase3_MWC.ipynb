# Actividad 1: Tipo de Parámetros

```python
# 1. Multiplicación (Posicionales)
def multiplicar(n1, n2):
    return n1 * n2

# 2. Bienvenida (Parámetros por defecto)
def bienvenida(nombre, mensaje="¡Bienvenido!"):
    return f"{mensaje}, {nombre}"

# 3. Promedio (*args - Tupla de argumentos)
def calcular_promedio(*notas):
    if not notas:
        return 0
    return sum(notas) / len(notas)

# 4. Información (**kwargs - Diccionario de argumentos)
def concatena_info(nombre, **datos):
    resumen = f"Usuario: {nombre}"
    for clave, valor in datos.items():
        resumen += f"\n - {clave.replace('_', ' ').capitalize()}: {valor}"
    return resumen

# --- ÁREA DE PRUEBAS ---
print("--- 1. Multiplicar ---")
print(f"Resultado: {multiplicar(7, 8)}")

print("\n--- 2. Bienvenida ---")
print(bienvenida("Lucas"))
print(bienvenida("Sofía", "Hola de nuevo"))

print("\n--- 3. Promedio ---")
print(f"Promedio de 3 notas: {calcular_promedio(10, 8, 9)}")
print(f"Promedio de 5 notas: {calcular_promedio(10, 10, 5, 7, 8)}")

print("\n--- 4. Concatena Info ---")
print(concatena_info("Mateo", edad=28, ciudad="Valencia", profesion="Diseñador"))
```

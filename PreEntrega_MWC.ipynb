## Etapa 1: Recopilación y Preparación de Datos (Clases 1 a 4)

```python
# Importar la librería Pandas
import pandas as pd
```

### Cargar Datasets

```python
def cargar_dataset(link):
  # Usamos regex para extraer el file_id del link
  file_id = link.split('/d/')[1].split('/')[0]

  # Con ese file_id generamos la url
  url = f"https://drive.google.com/uc?id={file_id}"

  # Finalmente lo importamos usando el método read_csv() con la url generada
  df = pd.read_csv(url)
  return df
```

#### Clientes

```python
# Link generado al compartir el archivo clientes.csv
link = "https://drive.google.com/file/d/1pyNXpqnOauR5tWJWj9f8C15Ygn5HmdWT/view?usp=drive_link"
df_clientes = cargar_dataset(link)
```

#### Ventas

```python
# Link generado al compartir el archivo ventas.csv
link = "https://drive.google.com/file/d/1dpvV5iOAKyD7QZX8gwAZVZIefWGpnTKM/view?usp=drive_link"
df_ventas = cargar_dataset(link)
```

#### Marketing

```python
# Link generado al compartir el archivo marketing.csv
link = "https://drive.google.com/file/d/1NV-ICitmc8HQjhNEuUcgKqvajoBbNBT5/view?usp=drive_link"
df_marketing = cargar_dataset(link)
```

### Análisis de Datos

```python
# Análisis de Datos
print("Clientes: --- Primeras filas del dataset ---")
display(df_clientes.head())
# Obtener información general (tipos de datos y memoria)
print("\n--- Información General ---")
df_clientes.info()
# Resumen estadístico de columnas numéricas
print("\n--- Resumen Estadístico ---")
display(df_clientes.describe())
# Ver las dimensiones del DataFrame (Filas, Columnas)
print(f"\nDimensiones del dataset: {df_clientes.shape}")

print("=====================================================================")

print("Ventas: --- Primeras filas del dataset ---")
display(df_ventas.head())
# Obtener información general (tipos de datos y memoria)
print("\n--- Información General ---")
df_ventas.info()
# Resumen estadístico de columnas numéricas
print("\n--- Resumen Estadístico ---")
display(df_ventas.describe())
# Ver las dimensiones del DataFrame (Filas, Columnas)
print(f"\nDimensiones del dataset: {df_ventas.shape}")

print("=====================================================================")

print("Marketing: --- Primeras filas del dataset ---")
display(df_marketing.head())
# Obtener información general (tipos de datos y memoria)
print("\n--- Información General ---")
df_marketing.info()
# Resumen estadístico de columnas numéricas
print("\n--- Resumen Estadístico ---")
display(df_marketing.describe())
# Ver las dimensiones del DataFrame (Filas, Columnas)
print(f"\nDimensiones del dataset: {df_marketing.shape}")

print("=====================================================================")
```

### Calidad de Datos

```python
# Calidad de Datos
print("Clientes: --- Calidad de Datos ---")
# Identificar valores nulos por columna
print("--- Valores Nulos por Columna ---")
nulos = df_clientes.isnull().sum()
print(nulos)
# Identificar filas duplicadas
duplicados = df_clientes.duplicated().sum()
print(f"\nCantidad de registros duplicados: {duplicados}")
# Documentación del estado inicial
print("\n--- Resumen de Calidad de Datos ---")
if nulos.sum() == 0 and duplicados == 0:
    print("El dataset no presenta valores nulos ni duplicados iniciales.")
else:
    print(f"Se detectaron {nulos.sum()} valores faltantes en total.")
    print(f"Se detectaron {duplicados} filas repetidas que requieren atención.")
print("=====================================================================")

print("Ventas: --- Calidad de Datos ---")
# Identificar valores nulos por columna
print("--- Valores Nulos por Columna ---")
nulos = df_ventas.isnull().sum()
print(nulos)
# Identificar filas duplicadas
duplicados = df_ventas.duplicated().sum()
print(f"\nCantidad de registros duplicados: {duplicados}")
# Documentación del estado inicial
print("\n--- Resumen de Calidad de Datos ---")
if nulos.sum() == 0 and duplicados == 0:
    print("El dataset no presenta valores nulos ni duplicados iniciales.")
else:
    print(f"Se detectaron {nulos.sum()} valores faltantes en total.")
    print(f"Se detectaron {duplicados} filas repetidas que requieren atención.")
print("=====================================================================")

print("Marketing: --- Calidad de Datos ---")
# Identificar valores nulos por columna
print("--- Valores Nulos por Columna ---")
nulos = df_marketing.isnull().sum()
print(nulos)
# Identificar filas duplicadas
duplicados = df_marketing.duplicated().sum()
print(f"\nCantidad de registros duplicados: {duplicados}")
# Documentación del estado inicial
print("\n--- Resumen de Calidad de Datos ---")
if nulos.sum() == 0 and duplicados == 0:
    print("El dataset no presenta valores nulos ni duplicados iniciales.")
else:
    print(f"Se detectaron {nulos.sum()} valores faltantes en total.")
    print(f"Se detectaron {duplicados} filas repetidas que requieren atención.")
print("=====================================================================")

```
